Importing our excel to a pandas dataframe and cleaning it out

```python
import pandas as pd

rt_df = pd.read_excel("RelationshipTweets_Labeled.xlsx", 0)

#Cleans out any empty data
rt_df = rt_df.loc[:, ~rt_df.columns.str.contains('^Unnamed')] # removes extra columns
rt_df = rt_df.dropna(how="all")  # removes rows with all NaN values
rt_df
```

Let's start creating our training dataset. We need:
    * tweet itself
    * comments
    * relationship

```python
trn_df = rt_df[["tweet", "Comments", "Relationship"]]
trn_df
```
